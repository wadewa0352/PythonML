```python
# Setup Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1111)
```

```python
# Load dataset
path = "/home/williamwade/"
file = "Phishing.csv"
phising_dataframe = pd.read_csv(path + file)
```

```python
#1. Get distribution of websites from the result column
results = phising_dataframe["Result"].value_counts().to_dict()

#Form the results for pandas
distribution = {'0': [ 1, results[1]], '1': [-1, results[-1]]}

#Make the keys the row value and add the column labels
df = pd.DataFrame.from_dict(distribution, orient='index',
                       columns=['Class', 'Num_Observations'])

df
```

```python
#2. Create a bar plot
ax = df.plot.barh(x='Class', y='Num_Observations', rot=0, title="Class distribution of the websites", color=['C0','C1'], figsize=(15,5))
ax.set_xlabel("Number of Observations")
ax.legend().remove()

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=15, color='dimgrey')

# invert for largest on top 
ax.invert_yaxis()
```
